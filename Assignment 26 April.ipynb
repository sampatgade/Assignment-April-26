```python
import pandas as pd

# Specify column names for the dataset
column_names = [
    "Wine Class",
    "Alcohol",
    "Malic Acid",
    "Ash",
    "Alcalinity of Ash",
    "Magnesium",
    "Total Phenols",
    "Flavanoids",
    "Nonflavanoid Phenols",
    "Proanthocyanins",
    "Color Intensity",
    "Hue",
    "OD280/OD315 of Diluted Wines",
    "Proline"
]

# Load the dataset into a DataFrame
data = pd.read_csv("wine.data", names=column_names)




```

```python
# Split the dataset into features and target variable
features = data.drop("Wine Class", axis=1)  # Drop the target variable column
target = data["Wine Class"]

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
missing_value = data.isnull().sum()

features_filled = features.fillna(features.mean())
```

```python
from sklearn.model_selection import train_test_split

# Assuming you have defined X and y before this step
X_train, X_test, y_train, y_test = train_test_split(scaled_features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
pca
```

```python
X_train=pca.fit_transform(X_train)
X_train
```

```python
X_test=pca.transform(X_test)
X_test
```

```python
import matplotlib.pyplot as plt
pca = PCA(n_components=2)  # Retain 2 principal components
principal_components = pca.fit_transform(scaled_features)
pca_df = pd.DataFrame(data=principal_components, columns=["PC1", "PC2"])
plt.figure(figsize=(10, 6))
plt.scatter(pca_df["PC1"], pca_df["PC2"], c=target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Scatter Plot')
plt.colorbar(label='Wine Class')
plt.grid()
plt.show()
```
